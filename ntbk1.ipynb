```python
df= pd.read_csv('data.csv')
```

```python
df1=df.replace(0,np.nan).dropna(axis=0).reset_index(drop=True)
df1
```

```python
traits = ['O','C','E','A','N']

for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1[traits]
```

```python
frames = ['O','C','E','A','N']

for frame in frames:
    frame_cols = sorted([col for col in df.columns if frame in col])
    df[frame] = df[frame_cols].mean(axis=1)
df[frames]
```

```python
traits = ['O']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['O_trait']=df1[traits]
df1['O_cat']=df1['O_trait'] > 25


traits = ['C']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['C_trait']=df1[traits]
df1['C_cat']=df1['C_trait'] > 25

traits = ['E']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['E_trait']=df1[traits]
df1['E_cat']=df1['E_trait'] > 25

traits = ['A']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['A_trait']=df1[traits]
df1['A_cat']=df1['A_trait'] > 25

traits = ['N']

for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['N_trait']=df1[traits]
df1['N_cat']=df1['N_trait'] >= 25



```
