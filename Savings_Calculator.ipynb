```python
from tkinter import *       #importing all the functions of tkinter library
root = Tk()                 #create a GUI Window

#creating global variable
variable_1=StringVar(root)
variable_2=StringVar(root)
variable_3=StringVar(root)

#vairable initialisation
variable_1.set("Expense")
variable_2.set("Expense")
variable_3.set("Expense")

#using Label Tkinter Widget Class to create GUI
if __name__ == "__main__" :
    root.configure(background= 'LavenderBlush2')           #Set the background colour
    root.geometry("400x200")                           #Set the dimension of the GUI window   
    
    #creating Labels
    headlabel=Label(root,text="Let's See How Much Have You Saved Today? :D" , fg="White" , bg="Black")
    label_5=Label(root,text="Salary: " , fg="Black" , bg="LavenderBlush2")
    label_1=Label(root,text="Food: " , fg="Black" , bg="LavenderBlush2")
    label_2=Label(root,text="Travel: " , fg="Black" , bg="LavenderBlush2")
    label_3=Label(root,text="Miscellaneous: " , fg="Black" , bg="LavenderBlush2")
    label_4=Label(root,text="Total Amount: " , fg="Black" , bg="LavenderBlush2")
    
    #placing the widgest using grid method
    headlabel.grid(row=0,column=1)
    label_1.grid(row=2,column=0)
    label_2.grid(row=3,column=0)
    label_3.grid(row=4,column=0)
    label_4.grid(row=6,column=0)
    label_5.grid(row=1,column=0)
    
    #creating entry text box
    Amount1_field=Entry(root)
    Amount2_field=Entry(root)
    Amount3_field=Entry(root)
    Amount5_field=Entry(root)
    
    Amount5_field.grid(row=1,column=1,ipadx="25")
    Amount1_field.grid(row=2,column=1,ipadx="25")
    Amount2_field.grid(row=3,column=1,ipadx="25")
    Amount3_field.grid(row=4,column=1,ipadx="25")
    
    answer_label=Label(root,text="Total")
    answer_label.grid(row=6,column=1)
    
    #creating buttons
    
    button1=Button(root,text="Calculate the expenses" , bg="orchid1" ,fg="black", command="callback")
    button1.grid(row=5, column=1)
    
    button2=Button(root,text="Clear" , bg="orchid1" ,fg="black", command="clear_all")
    button2.grid(row=7, column=1)
    
    root.mainloop()
    
    
    #callback function
def callback():
    num4=float(Amount5_field.get())
    num1=float(Amount1_field.get())
    num2=float(Amount2_field.get())
    num3=float(Amount3_field.get())
    
    answer= num4-(num1 + num2 + num3)
    answer_label.configure(text=answer)
    
    
    #clear_all function
    
def clear_all():
    Amount1_field.delete(0,END)
    Amount2_field.delete(0,END)
    Amount3_field.delete(0,END)
    
```
