```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('data.csv')
```

```python
df1=df.replace(0,np.nan).dropna(axis=0).reset_index(drop=True)
df1
```

```python
traits = ['O','C','E','A','N']

for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1[traits]
```

```python
frames = ['O','C','E','A','N']

for frame in frames:
    frame_cols = sorted([col for col in df.columns if frame in col])
    df[frame] = df[frame_cols].mean(axis=1)
df[frames]
df
```

```python
traits = ['O']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['O_trait']=df1[traits]
df1['O_cat']=df1['O_trait'] >=20

traits = ['C']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['C_trait']=df1[traits]
df1['C_cat']=df1['C_trait'] >= 20

traits = ['E']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['E_trait']=df1[traits]
df1['E_cat']=df1['E_trait'] >= 20

traits = ['A']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['A_trait']=df1[traits]
df1['A_cat']=df1['A_trait'] >=20

traits = ['N']
for trait in traits:
    trait_cols = sorted([col for col in df.columns if trait in col])
    df1[trait] = df1[trait_cols].sum(axis=1)
df1['N_trait']=df1[traits]
df1['N_cat']=df1['N_trait'] 

df1
```

```python
df2 = df1.drop(['O_trait','C_trait','E_trait','A_trait','N_trait'],axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
df2['country_cat'] = labelencoder.fit_transform(df1['country'])
df2 = df2.drop(['country'],axis = 1)
```

```python
df2.dtypes
```

```python
import scipy
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X = df2.iloc[:,0:19]  
y = df2.iloc[:,-1].values
```

```python
from sklearn.datasets import make_multilabel_classification

# this will generate a random multi-label dataset
X, y = make_multilabel_classification(sparse = True, n_labels = 20,
return_indicator = 'sparse', allow_unlabeled = False)
```

```python
classifier = BinaryRelevance(GaussianNB())

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

classifier.fit(X_train, y_train)

predictions = classifier.predict(X_test)
```

```python
print(predictions)
```

```python
# using classifier chains
from skmultilearn.problem_transform import ClassifierChain
from sklearn.naive_bayes import GaussianNB

# initialize classifier chains multi-label classifier
# with a gaussian naive bayes base classifier
classifier = ClassifierChain(GaussianNB())

# train
classifier.fit(X_train, y_train)

# predict
predictions = classifier.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
