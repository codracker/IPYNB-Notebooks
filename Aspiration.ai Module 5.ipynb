# Problem Statement 5.1 
For your chosen stock, calculate the mean daily return and daily standard deviation of returns, and then just annualise them to get mean expected annual return and volatility of that single stock. ( annual mean = daily mean 252 , annual stdev = daily stdev sqrt(252) )

```python
import pandas as pd
itdc_stock_data = pd.read_csv('itdc.csv')
itdc_stock_data
```

```python
import numpy as np
itdc_stock_data['Daily Return'] = (itdc_stock_data['Close Price']).pct_change() 
itdc_stock_data['Daily Return'] = itdc_stock_data['Daily Return'].replace([np.inf, -np.inf], np.nan)
itdc_stock_data = itdc_stock_data.dropna()
print("Mean Daily Return")
itdc_stock_data['Daily Return'].mean()
```

```python
itdc_stock_data['Daily Standard Deviation'] = (itdc_stock_data['Close Price']).pct_change() 
itdc_stock_data['Daily Standard Deviation'] = itdc_stock_data['Daily Standard Deviation'].replace([np.inf, -np.inf], np.nan)
itdc_stock_data = itdc_stock_data.dropna()
print("Daily Standard Deviation")
itdc_stock_data['Daily Standard Deviation'].std()
```

```python
import math
annual_stdev = 0.013515857199631161 * math.sqrt(252)
print("Annual Standard Deviation: "+ str(annual_stdev))
```
