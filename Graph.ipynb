```python
dict = {}
class graph:
    def main_menu(self):
        print("1.Add a node")
        print("2.Delete a node")
        print("3.Add an Edge")
        print("4.Delete an Edge")
        print("5.BFS")
        print("6.DFS")
        print("7.Print the Graph")
        print("8.Quit")
        choice = int(input("Enter Your Choice of operation:"))
        if choice == 1:
            self.Add_a_node()
        elif choice == 2:
            self.delete_a_node()
        elif choice == 3:
            self.add_an_edge()
        elif choice == 4:
            self.delete_an_edge()
        elif choice == 5:
            self.BFS()
        elif choice == 6:
            self.DFS()
        elif choice == 7:
            self.print_graph()
        elif choice == 8:
            quit()
        else :
            print("Wrong Choice,Enter a valid option!")
            break
    def add_a_node(self):
        key = input("Enter the node to insert")
        if key not in dict:
            dict[key]=[]
            print("Sucessful Insertion!")
        else:
            print("Try Again!")
            self.main_menu()
    
    def delete_a_node(self):
        key = input("Enter the node")
        if key in dict:
            dict.pop(key)
            print("Sucessful Deletion!")
        else:
            print("Unsucessful !")
            self.print_menu()

    def add_an_edge(self):
        n=int(input("Enter the source vertex: "))
        i=int(input("Enter the destination vertex: "))
        if n in dict:
            if i in dict:
                dict[n].append(i)
                print("Sucessful Addition!")
            else:
                print("Unsucessful, Try Again!")
        else:
            print("Sorry,Try Again!")
            self.print_menu()
    
    def delete_an_edge(self):
        n=int(input("Enter the source node"))
        i=int(input("Enter the destination node"))
        if n in dict:
            if i in dict:
                dict[n].remove(i)
                print("Sucessful Deletion!")
            else:
                print("Oops,Sorry!")

        else:
            print("Unsucessful!")
            self.print_menu()
        
    def print_graph(self):
        print(dict)
        self.main_menu()

                  
g = graph()
g.main_menu()
```
